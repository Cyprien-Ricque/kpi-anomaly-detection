```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import seaborn as sns
```

```python
result_file = '../predict.csv'
truth_file = '../ground_truth.hdf'
results_ref = '../predict_reference.csv'
```

```python
results = pd.read_csv(result_file)
results_ref = pd.read_csv(results_ref)
```

```python
nb_max_to_plot = 5

print(f'Plot {nb_max_to_plot} KPIs')

for kpi_id in results['KPI ID'].unique()[:nb_max_to_plot]:
    figure = plt.figure(figsize=(20, 2))

    data = results_ref[results_ref['KPI ID'] == kpi_id]
    sns.lineplot(data=data, x='timestamp', y='predict', color='blue', label='reference', alpha=0.1, figure=figure)

    data = results[results['KPI ID'] == kpi_id]
    sns.lineplot(data=data, x='timestamp', y='predict', figure=figure, alpha=0.1, color='r', label='mine')

    plt.title(f'results {kpi_id}')
    plt.legend()
    plt.show()
```
